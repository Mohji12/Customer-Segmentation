### Model Building

```python
dff = df.copy()
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
scaled_df = scalar.fit_transform(dff)
scaled_df
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

```python
from sklearn.cluster import KMeans

inertia = []
range_val = range(1,11)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show()
```

```python
from kneed import KneeLocator

kl = KneeLocator(range(1,11),inertia,curve="convex",direction="decreasing")

kl.elbow
```

```python
from sklearn.cluster import KMeans

kmeans_model=KMeans(n_clusters=3, random_state=0, init = 'k-means++')
kmeans_model.fit_predict(scaled_df)

#pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
#pca_df_kmeans
pca_df['Cluster'] = kmeans_model.labels_
pca_df
```

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="Cluster",data=pca_df,palette=['red','green','blue','black','yellow'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score, silhouette_score

print(silhouette_score(pca_df, kmeans_model.labels_))
print(calinski_harabasz_score(pca_df, kmeans_model.labels_))
print(davies_bouldin_score(pca_df, kmeans_model.labels_))
```

### Testing with Classification Models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB

models = {
    "RandomForest": RandomForestClassifier(criterion="entropy"),
    "DecisionTreeClassifier": DecisionTreeClassifier(criterion="entropy"),
    "LogisticRegression": LogisticRegression(),
    "NaiveBayes": GaussianNB(),  # Gaussian Naive Bayes
    "GradientBoosting": GradientBoostingClassifier()
}
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def evaluate_model(X_train, Y_train, X_test, Y_test, models):
    reports = {}
    for name, model in models.items():
        model.fit(X_train, Y_train)
        Y_test_pred = model.predict(X_test)
        accuracy = accuracy_score(Y_test, Y_test_pred)
        class_report = classification_report(Y_test, Y_test_pred)
        conf_matrix = confusion_matrix(Y_test, Y_test_pred)
        reports[name] = {
            "Accuracy": accuracy,
            "Classification Report": class_report,
            "Confusion Matrix": conf_matrix
        }
    return reports

reports = evaluate_model(X_train, Y_train, X_test, Y_test, models)

for name, metrics in reports.items():
    print(f"Model: {name}")
    print(f"Accuracy: {metrics['Accuracy']:.4f}")
    print(f"\nClassification Report:\n{metrics['Classification Report']}")
    print(f"\nConfusion Matrix:\n{metrics['Confusion Matrix']}")
    print()
```

Conclusion:

- Based on the comprehensive evaluation, Gradient Boosting emerges as the top-performing model, with the highest accuracy and robust performance across all evaluation metrics. It demonstrates superior classification performance and minimal misclassifications compared to other models.
- Random Forest also performs well and can be considered a viable alternative, offering high accuracy and balanced performance across all classes.
- Decision Tree Classifier, Logistic Regression, and Naive Bayes, while still effective, exhibit slightly lower performance metrics compared to Random Forest and Gradient Boosting.
- Overall, Gradient Boosting is recommended as the preferred model for this classification task, followed by Random Forest as a strong alternative
